### Visualizar relaciones entre datos

Primero, importamos las librerías necesarias y cargamos el conjunto de datos:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#Cargamos el conjunto de datos
planets = sb.load_dataset("planets")
planets.head(5)
```

El análisis estadístico es un proceso de comprensión de cómo las variables de un conjunto de datos se relacionan entre sí y cómo esas relaciones dependen de otras variables. La visualización puede ser un componente central de este proceso porque, cuando los datos se visualizan correctamente, podemos observar tendencias y patrones que indican una relación.

Usaremos 3 funciones de Seaborn:
* <code>relplot()</code>
* <code>scatterplot()</code>
* <code>lineplot()</code>

<b>Relacionar variables con diagramas de dispersión</b>

El diagrama de dispersión representa la distribución conjunta de dos variables mediante una nube de puntos, donde cada punto representa una observación en el conjunto de datos. Esta descripción permite al ojo inferir una cantidad sustancial de información sobre si existe alguna relación significativa entre ellos. Hay varias formas de dibujar un diagrama de dispersión en seaborn. 

La más básica, que debe usarse cuando ambas variables son numéricas, es la función <code>scatterplot ()</code>. <code>scatterplot ()</code> es el tipo por defecto en <code>relplot()</code>, puede especificarse también usando el parámetro kind="scatter".

```python
sb.relplot(x="distance", y="mass", hue="method", data=planets)
```

Si bien los puntos se representan en dos dimensiones, se puede agregar otra dimensión al gráfico coloreando los puntos de acuerdo con una tercera variable, como si fuese una leyenda. En seaborn, esto se denomina uso de una "semántica de tono", porque el color del punto adquiere significado.

Además, para enfatizar la diferencia entre las clases y mejorar la accesibilidad, puede usar un estilo de marcador diferente para cada clase. Con este variable semántica, podemos representar cuatro variables cambiando el tono y el estilo de cada punto de forma independiente.

```python
sb.relplot(x="distance", y="mass", hue="year", style="number", data=planets)
```

Destacar que si la variable que usamos con el parámetro hue es <b>categórica</b>, se usará el tono semántico por defecto. En cambio, si la variable es <b>numérica</b>, se aplicará un tono secuencial. También es posible personalizar la paleta de colores usando la interfaz de string para cubehelix_palette ():

```python
sb.relplot(x="distance", y="mass", hue="year", palette="ch:r=-.5,l=.75", data=planets)
```

El tercer tipo de variable semántica cambia el tamaño de cada punto, usando <code>size</code>. El tamaño de cada punto se puede modificar usando <code>sizes</code>

```python
sb.relplot(x="distance", y="mass", hue="year", size="year", sizes=(15,150), data=planets)
```

<b>Enfatizando la continuidad con diagramas de líneas</b>

Los gráficos de dispersión son muy eficaces, pero no existe un tipo de visualización óptimo universalmente. En cambio, la representación visual debe adaptarse a las características específicas del conjunto de datos y a la pregunta que está tratando de responder con la trama.

Con algunos conjuntos de datos, es posible que desee comprender los cambios en una variable en función del tiempo o una variable continua similar. En esta situación, una buena opción es dibujar una gráfica lineal. En seaborn, esto se puede lograr con la función <code>lineplot()</code>, ya sea directamente o con <code>relplot()</code> configurando kind = "line":

```python
a= sb.relplot(x="distance", y="mass", kind="line", data=planets)
a.fig.autofmt_xdate()
```

Debido a que <code>lineplot()</code> asume que la mayoría de las veces está tratando de dibujar y como una función de x, el comportamiento predeterminado es ordenar los datos por los valores de x antes de graficar. Sin embargo, esto se puede desactivar con el parámetro sort=False:

```python
a= sb.relplot(x="distance", y="mass", kind="line", sort=False, data=planets)
a.fig.autofmt_xdate()
```
